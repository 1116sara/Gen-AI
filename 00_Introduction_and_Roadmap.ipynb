## 🧠 Core Mental Models (Build These First)

### Why Transformers Replaced RNNs
```
RNN/LSTM Problem 1 → Vanishing gradients (forget early tokens during backprop)
RNN/LSTM Problem 2 → Sequential computation (cannot parallelize)
Transformer Solution → Self-Attention (every token sees every token, in parallel)
```

### The Golden Rule of LLM Engineering
> The model predicts the next token based on ALL tokens in context.
> Better context = better predictions. Always.

### The Three Core Tradeoffs You'll Make Every Day
```
Cost     vs  Quality      → more tokens = better output = more expensive
Speed    vs  Accuracy     → bigger model = more accurate = slower
Memory   vs  Relevance    → more history = more context = more tokens
```
