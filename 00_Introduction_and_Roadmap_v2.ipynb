## Core Mental Models

Why Transformers replaced RNNs. RNNs had vanishing gradients and sequential computation. Transformers introduced self-attention — every token sees every token simultaneously in parallel.

The golden rule of LLM engineering. The model predicts the next token based on all tokens in context. Better context equals better predictions. Always.

The three core tradeoffs made every day.
Cost vs Quality — more tokens means better output means more expensive.
Speed vs Accuracy — bigger model means more accurate means slower.
Memory vs Relevance — more history means more context means more tokens.

The persistence pattern. Build index once, save to disk, load forever. Never recompute what you have already computed.

Separation of concerns. Function finds and filters results, returns clean list. Caller checks if list is empty and handles display. Never mix printing and data building.
