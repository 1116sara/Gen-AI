Correct. And that's actually the full answer.

The system message is **not architecturally special**. It's just text. It gets placed at the beginning of the context window, before the user message, and the model sees it all as one continuous sequence of tokens.

The format looks like this internally:

```
[SYSTEM TOKENS] You are an expert Python teacher...
[USER TOKENS] What is a token in LLMs?
[ASSISTANT TOKENS] ← model predicts from here
```

The model was fine-tuned to respect the pattern — whatever appears in the system position shapes how it completes the assistant section. But mechanically it's all just tokens in a sequence.

This has one very important practical implication that most beginners never think about:

* The system message consumes tokens from your context window just like everything else.

So if you write a 2000 token system prompt, you've already spent 2000 tokens before the user says a single word.
